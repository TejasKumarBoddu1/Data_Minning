```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data={'categor':['a','b','c','a','b','c']}
df=pd.DataFrame(data)
```

```python
df
```

```python
#label encoding
Label_Encoder=LabelEncoder()
df['Label_Encoder']=Label_Encoder.fit_transform(df['categor'])
```

```python
#one_hot_encoding
one_hot_encoder=pd.get_dummies(df['categor'],prefix='categor')
```

```python
df=pd.concat([df,one_hot_encoder],axis=1)
```

```python
df
```
